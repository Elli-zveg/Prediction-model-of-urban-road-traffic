Удаляем дни, в которых не 24 часа

```python
dddf = df.groupby(['Month',	'Day', 'Day of Week'], 
                  as_index=False)['Hour'].count()
dddf = dddf.loc[(dddf['Hour']!= 24)]

df_second = df
for i in range (0, len(dddf)):
  m = dddf.iloc[i]['Month']
  d = dddf.iloc[i]['Day']
  w = dddf.iloc[i]['Day of Week']
  df_second = df_second.loc[((df['Month'] != m)|
                             (df['Day'] != d)|(df['Day of Week'] != w))]
```

```python
import matplotlib # Для графиков
import matplotlib.pyplot as plt

import numpy as np 
import pandas as pd # Для датафреймов
import seaborn as sns # Для графиков

import sklearn # Для стандартизации
from sklearn import preprocessing
```

```python
df_s = df_second.drop(columns=['Month',	'Day',	'Hour', 'Day of Week'])
df_s
df_s.to_excel('drive/MyDrive/code/df_s.xlsx', index=False)
```

```python
fig, (ax1) = plt.subplots(ncols = 1, figsize = (10, 8))
ax1.set_title('Оригинальные распределения')

# kdeplot() (KDE – оценка плотности ядра) – специальный метод для графиков распределений
#sns.kdeplot(df_s['Volume'], ax = ax1)
sns.kdeplot(df_s['Occupancy'], ax = ax1)
```

```python
# Инициализируем стандартизатор
s_scaler = preprocessing.StandardScaler()

# Копируем исходный датасет
df_s_ = s_scaler.fit_transform(df_s)

# Копируем названия столбцов, которые теряются при использовании fit_transform()
col_names = list(df_s.columns)

# Преобразуем промежуточный датасет в полноценный датафрейм для визуализации
df_s_ = pd.DataFrame(df_s_, columns = col_names)

# Расположим распределения на одном полотне
fig, (ax1) = plt.subplots(ncols = 1, figsize = (10, 8))

# Зададим название графика
ax1.set_title('После стандартизации')

#sns.kdeplot(df_s_['Volume'], ax = ax1)
sns.kdeplot(df_s_['Occupancy'], ax = ax1)
```

В связи с большой корреляцией Пирсона, берем один элемент для дальнейшего прогнозирования https://www.statisticshowto.com/probability-and-statistics/correlation-coefficient-formula/

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Выберем встретившиеся числовые типы
numerics = ['Occupancy']

# Создадим копию датасета
newdf = df_second['Occupancy']
newdf

# Инициализация нормализатора
scaler = MinMaxScaler()

# Передача датасета и преобразование
scaler.fit(newdf)
scaled_features = scaler.transform(newdf)
scaled_features
inversed = scaler.inverse_transform(scaled_features)
inversed

# Конвертация в табличный формат.
df_MinMax = pd.DataFrame(data = scaled_features, 
columns = ['Month',	'Day',	'Hour', 'Day of Week', 'Occupancy'])

df_MinMax
```
